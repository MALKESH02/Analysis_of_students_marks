Problem3
Standard deviation for given subject.

```python
#Mean of all the subjects.
mean_phy=sum(phy)/len(phy)
mean_chem=sum(chem)/len(chem)
mean_math=sum(math)/len(math)

#Standard deviation of all the subjects.
def Standar_deviation(phy,chem,math):
    x=0
    y=0
    z=0
    for i in range (len(phy)):
        x = x + (phy[i]-mean_phy)**2/len(phy)
        y = y + (chem[i]-mean_chem)**2/len(chem)
        z = z + (math[i]-mean_math)**2/len(math)
    
    standard_deviation1=x**0.5
    standard_deviation2=y**0.5
    standard_deviation3=z**0.5

    print('standard deviation for subject Physcis is -:',standard_deviation1)
    print('standard deviation for subject Chemistry is -:',standard_deviation2)
    print('standard deviation for subject Maths is -:',standard_deviation2)

    standard_deviation=[standard_deviation1,standard_deviation2,standard_deviation3]

    Subjects=['Physcis','Math','Chem']

    fig=plt.figure(figsize=(10,5))


    plt.bar(Subjects,standard_deviation,color='orange',edgecolor='black',width=0.4)

    plt.xlabel('Subjects')
    plt.ylabel('Standard deviation')
    plt.title('Standard deviation of given Subjects')
    plt.grid(axis = 'y',color='black',linestyle = '--')
    plt.show()
Standar_deviation(phy,chem,math)
```

Conclusion3
1) Physcis differ 12.575 units from the mean value of the data.
1) Chemistry & Maths differ 14.129 units from the mean value of the data.


Problem4
Is there any corelation between chemistry and maths? as standard devistion of chemistry and Maths is same then there might be any corelation between them.

H0=There is corelation between chemistry and maths.
H1=There is no co-relation between chemistry and maths.

```python

mean_chem=sum(chem)/len(chem)
mean_math=sum(math)/len(math)

chem_difference=[]
square_of_difference_chem=[]
sum_of_square_chem=0
for i in range (len(chem)):
        difference=chem[i]-mean_chem
        chem_difference.append(difference)
        differencesquarex = chem_difference[i]**2
        square_of_difference_chem.append(differencesquarex)
        sum_of_square_chem=sum_of_square_chem+square_of_difference_chem[i]
math_difference=[]
square_of_difference_math=[]
sum_of_square_math=0
for i in range (len(math)):
        difference1=math[i]-mean_math
        math_difference.append(difference1)
        differencesquare = math_difference[i]**2
        square_of_difference_math.append(differencesquare)
        sum_of_square_math=sum_of_square_math+square_of_difference_math[i]
product=[]

for i in range (len(math)):
              covariance1= (math_difference[i]*chem_difference[i])
              product.append(covariance1)
sum4=0

for i in range (len(math)):
    sum4= sum4 + product[i]
              
Numerator=sum4

denominator=(sum_of_square_math*sum_of_square_chem)**0.5

corelation=Numerator/denominator

print('corelation between math and chem is -:',corelation)
```

Conclusion3-:

standard deviation is the same then also there is very less correlation between math and chemistry, it is not required for students to perform well in arithmetic to perform well in chemistry or vice versa.
